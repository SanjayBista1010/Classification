```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('adult.csv')
df
```

```python
df = df.replace('?', np.nan)
df
```

```python
for col in ['workclass', 'occupation', 'native-country']:
    df[col] = df[col].fillna(df[col].mode()[0])
df.info()
```

```python
plt.figure(figsize=(8,8))
sns.countplot(data=df, x='marital-status', hue='income')
plt.show()

sns.countplot(data=df, x='relationship', hue='income')
plt.show()

sns.countplot(data=df, x='occupation', hue='income')
plt.show()

sns.countplot(data=df, x='gender', hue='income')
plt.show()

sns.countplot(data=df, x='race', hue='income')
plt.show()
```

```python
df['workclass'].unique()
```

```python
workclass_mapping = {
    'Never-worked': 0,
    'Without-pay': 1,
    'Self-emp-not-inc': 2,
    'Self-emp-inc': 3,
    'Private': 4,
    'Local-gov': 5,
    'State-gov': 6,
    'Federal-gov': 7
}

df['workclass'] = df['workclass'].map(workclass_mapping)
```

```python
df['education'].unique()
```

```python
education_mapping = {
    'Preschool': 0,
    '1st-4th': 0,
    '5th-6th': 0,
    '7th-8th': 0,
    '9th': 0,
    '10th': 0,
    '11th': 1,
    '12th': 1,
    'HS-grad': 2,
    'Some-college': 2,
    'Assoc-acdm': 2,
    'Assoc-voc': 2,
    'Bachelors': 3,
    'Masters': 4,
    'Doctorate': 5,
    'Prof-school': 2
}

df['education'] = df['education'].map(education_mapping)
```

```python
df['marital-status'].unique()
```

```python
mapping = {
    'Married-civ-spouse': 'Married',
    'Married-spouse-absent': 'Married',
    'Married-AF-spouse': 'Married',
    'Divorced': 'Divorced',
    'Never-married': 'Unmarried',
    'Separated': 'Unmarried',
    'Widowed': 'Unmarried'
}

df['marital-status'] = df['marital-status'].map(mapping)
```

```python
df = pd.get_dummies(df, columns=['marital-status','occupation','race','relationship'], dtype=int)
```

```python
df['gender'].unique()
```

```python
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1})
```

```python
df['native-country'].unique()
```

```python
freq_encoding = df['native-country'].value_counts() / len(df)
df['native-country_freq'] = df['native-country'].map(freq_encoding)
```

```python
df
```

```python
df['income'] = df['income'].map({'<=50K': 0, '>50K': 1})
```

```python
df.info()
```

```python
encoded_df = df.select_dtypes(include=['int64', 'float64']).copy()
encoded_df
```

```python
y = encoded_df['income']
y
```

```python
x = encoded_df.drop('income', axis=1)
x
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, stratify=y)

sm = SMOTE(random_state=42)
x_sampled, y_sampled = sm.fit_resample(x_train, y_train)
print(y_sampled.value_counts())
y_train.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_sampled_scaled = scaler.fit_transform(x_sampled)
x_test_scaled = scaler.transform(x_test)

x_sampled_scaled, x_test_scaled
```

```python
correlation = encoded_df.corr()['income'].drop('income')
correlation.sort_values(ascending=False)
```

```python
features = [
    'marital-status_Married',
    'relationship_Husband',
    'educational-num',
    'education',
    'age',
    'hours-per-week',
    'capital-gain',
    'occupation_Exec-managerial',
    'capital-loss',
    'relationship_Wife',
    'occupation_Prof-specialty',
    'race_White',
    'marital-status_Divorced',
    'relationship_Unmarried',
    'occupation_Other-service',
    'relationship_Not-in-family',
    'gender',
    'relationship_Own-child',
    'marital-status_Unmarried'
]

for col in features:
    sns.countplot(data=encoded_df, x=col, hue='income')
    plt.title(f'Income by {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold, StratifiedKFold, cross_val_score,cross_val_predict
from sklearn.metrics import accuracy_score, classification_report

logistic_model = LogisticRegression()
kf = KFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(logistic_model, x_sampled_scaled, y_sampled, cv=kf, scoring='accuracy')

print(f"Accuracy scores for each fold: {scores}")
print(f"Mean Accuracy: {np.mean(scores):.4f}")
print(f"Standard Deviation: {np.std(scores):.4f}")


x_sampled_scaled = pd.DataFrame(x_sampled_scaled)
y_train = pd.Series(y_train)
fold = 1
for train_idx, val_idx in kf.split(x_sampled_scaled, y_sampled):
    logistic_model.fit(x_sampled_scaled.iloc[train_idx], y_sampled.iloc[train_idx])
    y_val_pred = logistic_model.predict(x_sampled_scaled.iloc[val_idx])
    print(f"\nFold {fold} Classification Report:")
    print(classification_report(y_sampled.iloc[val_idx], y_val_pred))
    fold += 1

logistic_model.fit(x_sampled_scaled, y_sampled)
y_pred = logistic_model.predict(x_test_scaled)
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
